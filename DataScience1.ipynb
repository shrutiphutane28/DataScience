```python
dataframe=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',names=cols,na_values='?')
dataframe
```

```python
dataframe.isnull().sum()
```

```python
dataframe.rename(columns={'horsepower':'Power(hp)'}, inplace='True')
dataframe.columns
```

```python
columns_to_fill = ['normalized-losses', 'stroke', 'bore', 'Power(hp)', 'peak-rpm']
for column in columns_to_fill:
    dataframe[column].replace(np.nan, dataframe[column].mean(), inplace=True)
dataframe
```

```python
dataframe['num-of-doors'].replace(np.NaN,"four",inplace=True)
dataframe
```

```python
dataframe.dropna(subset=['price'],inplace=True)
dataframe
```

```python
dataframe.dtypes
```

```python
dataframe['city-L/100km'] = 235/dataframe['city-mpg']
dataframe['highway-L/100km'] = 235/dataframe['highway-mpg']
dataframe.head(5)
```
